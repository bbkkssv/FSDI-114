```python
class Stack:

    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items) == 0

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.items.pop()

    def peek(self):
        if self.is_empty():
            return None
        return self.items[-1]

    def size(self):
        return len(self.items)
```

```python
def is_balanced(string): #This checks to see if the parentheses in the string are balanced (assuming you are looking for items in parentheses 1 = parentheses 2 )
    stack = Stack() # Calling previously created stack class
    for char in string: # This is itterating through each character in the string. It is looking for balanced parentheses and not necessarily the content of the parentheses.
        if char == "(":
            stack.push(char)
        elif char == ")": # Until the end of the string, if it finds a closing parentheses
            if stack.is_empty():
                return False
            stack.pop()
    return stack.is_empty()
    
# Instructor Examples:

print(is_balanced("(a + b) * (c + d)"))


print(is_balanced("(a + b * (c + d)"))
```

```python
# create a stack
stack1 = Stack()

# Push three items
stack1.push(5)
stack1.push(10)
stack1.push(15)

# Peek at the top item
print("Stack peek: ", stack1.peek())

# Pop the top item
print("Stack pop: ", stack1.pop())

# Check the size
print("Stack size: ", stack1.size())
```

```python
class Queue:

    def __init__(self):
        self.items = []

    # is_empty() method
    def is_empty(self):
        return len(self.items) == 0

    # enqueue() method - adds items to the back of the queue
    def enqueue(self, item):
        self.items.append(item)

    # dequeue() method - remove an item
    def dequeue(self):
        if self.is_empty():
            return None
        return self.items.pop(0)

    # size() method - returns the number of items
    def size(self):
        return len(self.items)
```

```python
q = Queue()


print("Empty?", q.is_empty())


q.enqueue(10)
q.enqueue(20)
q.enqueue(30)
print("After adding 10, 20, 30:", q.items)


print("Size:", q.size())


print("Empty?", q.is_empty())


removed = q.dequeue()
print("Dequeued:", removed)
print("After dequeue:", q.items)


print("Size:", q.size())
```
