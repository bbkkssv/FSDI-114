HOW TO FIND THE LARGEST NUMBER IN A LIST:

1. Check if the list is empty:
   - If the list has no numbers, stop and say "Error"

2. Set up a variable to track the largest number:
   - Take the first number from the list
   - Call this your "largest number so far"

3. Compare with all other numbers:
   - Go through each remaining number in the list one by one
   - For each number:
     * Ask: "Is this number bigger than my largest number so far?"
     * If YES: Replace "largest number so far" with this new number
     * If NO: Keep your current "largest number so far" and move to the next

4. When you've checked all numbers:
   - The "largest number so far" is now the actual largest number
   - Return this as your answer

END

```python
def find_largest(number):
    if not number:
        return "Error"
    
    largest_so_far = number[0]
    
    for num in number[1:]:
        if num > largest_so_far:
            largest_so_far = num
    
    return largest_so_far
```

```python
numbers = [12, 4, 27, 3, 188, 50, 7]
result = find_largest(numbers)
print("Largest number:", result)
```

```python
def count_even_numbers(nums):
    increment = 0
    for num in nums:
        if num % 2 == 0:
            increment += 1
    return increment

numbers = [3,5,27,20,6,3.5,60]
result = count_even_numbers(numbers)
print(result)
        
```
