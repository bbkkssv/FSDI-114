## Stacks
A stack is an ordered collection of elements where items are added and removed from one end, commonly referred to as the "top", The ordering principle is referred to as LIFO (Last In, First Out)

## Queues
A queue is an ordered collection of elements where items are added and removed from **opposite** ends, commonly referred to as the "front" and "back" or "rear" of the queue. The ordering principle is referred to as FIFO (First In, First Out) or "first come first served"

```python
# Challenge 1: Push and Pop

# 1. Create an empty stack
stack = []

# 2. Push the values "A", "B", and "C" onto the stack in that order
stack.append("A")
stack.append("B")
stack.append("C")

# 3. Pop the top value and store it in a variable called top
top = stack.pop()

# 4. Print the entire stack and print the value that was removed
print("Stack after first pop:", stack)
print("Removed value:", top)

# 5. Pop the next value from the stack and print it
removed = stack.pop()
print("Next removed value:", removed)

# 6. Push the value "D" onto the stack
stack.append("D")

# 7. Pop again and print the removed value
removed = stack.pop()
print("Removed value:", removed)

# 8. Print the entire stack
print("Final stack:", stack)


```

```python
stack4 = []
stack4.append("H")
stack4.append("E")
stack4.append("L")
stack4.append("L")
stack4.append("O")

reversed = ""
while len(stack4) > 0:
    reversed += stack4.pop()

print("Reversed hello:", reversed)
```
