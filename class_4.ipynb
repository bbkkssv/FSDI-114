```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        current = self.head
        while current.next is not None:
            current = current.next
        current.next = new_node

    # Challenge 3.1: print_list method
    def print_list(self):
        current = self.head

        if current is None:
            print("List is empty!")
            return

        while current is not None:
            print(current.data)
            current = current.next

    # Challenge 3.2: count_nodes method
    def count_nodes(self):
        counter = 0
        current_node = self.head
        while current_node is not None:
            counter += 1
            current_node = current_node.next
        return counter

    # Challenge: prepend method
    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Challenge: remove method
    def remove(self, value):
        # Remove the first node that has the given value.
        # If the list is empty, nothing to do
        if self.head is None:
            return

        # If the node to remove is the head
        if self.head.data == value:
            self.head = self.head.next
            return

        # Otherwise, search for the node before the one to remove
        current = self.head
        while current.next is not None:
            if current.next.data == value:
                # Skip the node with the matching value
                current.next = current.next.next
                return
            current = current.next

   
```

```python
linked_list = LinkedList()
linked_list.append(10)
linked_list.append(15)
linked_list.append(20)
linked_list.append(25)
linked_list.append(30)

linked_list.print_list()
print("Number of nodes:", linked_list.count_nodes())

linked_list.prepend(50)
linked_list.print_list()
print("Number of nodes:", linked_list.count_nodes())

linked_list.remove(25)
linked_list.print_list()
print("Number of nodes after remove:", linked_list.count_nodes())

linked_list.remove(55)
linked_list.print_list()
print("Number of nodes after remove:", linked_list.count_nodes())
```

```python
class Node:
    def __init__(self, data):
        self.previous = None
        self.data = data
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Challenge 3: append method
    def append(self, value):
        # Create a new node with the given value
        new_node = Node(value)
        
        # Check if the list is empty
        if self.head is None:
            # If it is empty, make the new node the head
            self.head = new_node
            # Stop
            return
        
        #Start traversing the list from the head
        current = self.head
        
        # Move forward until we reach the last node
        while current.next is not None:
            current = current.next
        
        # Link the last node to the new node
        current.next = new_node
        
        # Link the new node back to the previous last node
        new_node.previous = current

    # Challenge 4: print_forward method
    def print_forward(self):
        current = self.head
        
        if current is None:
            print("List is empty!")
            return
        
        while current is not None:
            print(current.data)
            current = current.next

   # Challenge 5: print_backward method
    def print_backward(self):
        current = self.head
        
        if current is None:
            print("List is empty!")
            return
        
        # Move to tail
        while current.next is not None:
            current = current.next
        
        # Walk backwards using prev
        while current is not None:
            print(current.data)
            current = current.previous

doubly_linked_list = DoublyLinkedList()
doubly_linked_list.append("A")
doubly_linked_list.append("B")
doubly_linked_list.append("C")  
doubly_linked_list.append("D")

print("Forward:")
doubly_linked_list.print_forward()
print("Backward:")
doubly_linked_list.print_backward()
```
